```python
import numpy as np
import pandas as pd
```

```python
data['Ram'] = data['Ram'].str.replace('GB', '').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg', '').astype('float32')
```

```python
data['Company'].value_counts()
```

```python
def add_company(company_name):
    if company_name =='Samsung' or company_name =='Mediacom' or company_name =='Razer' or company_name =='Microsoft' or company_name =='Vero' or company_name =='Xiaomi' or company_name =='Chuwi' or company_name =='Fujitsu' or company_name =='Google' or company_name =='LG' or company_name =='Huawei':
        return 'Other'
    else:
        return company_name

data['Company'] = data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
data.head()
```

```python
data['Product'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['TouchScreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head()
```

```python
data['Cpu'].value_counts()
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_Processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

data['cpu_name'] = data['cpu_name'].apply(set_Processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data.head(2)
```

```python
data['Gpu'].value_counts()
```

```python
data['Gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data['Gpu_name'].value_counts()
```

```python
data = data[data['Gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
data['OpSys'].value_counts()
```

```python
def set_OpSys(op):
    if op == 'Windows 10' or op =='Windows 7' or op =='Windows 10 S':
        return 'Windows'
    elif op == 'macOS' or op == 'Mac OS X':
        return 'Mac'
    elif op == 'Linux':
        return op
    else:
        return 'Other'

data['OpSys'] = data['OpSys'].apply(set_OpSys)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data =data.drop(columns=['laptop_ID', 'Product', 'Inches', 'ScreenResolution', 'Cpu', 'Gpu'])
```

```python
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.shape
```

```python
x =  data.drop('Price_euros', axis=1)
y =  data['Price_euros']
```

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, Y_train)
    acc = model.score(X_test, Y_test)
    print(str(model) + '--> ' + str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators': [10, 50, 100],
              'criterion': ['squared_error', 'absolute_error', 'poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(X_train, Y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, Y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
